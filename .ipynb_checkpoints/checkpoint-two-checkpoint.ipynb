```python
# get column name, null count and data type info
life_df.info()
```

```python
# summary of row values for each column
# checking to see if there are really no null or null-equivalent values
for col in life_df.columns:
    print(life_df[col].value_counts())
    print(life_df[col].describe())
    print()
```

```python
# checking for duplicates to see if a region is listed >1
print(life_df.duplicated().any())
```

```python
# the rows for each column that have 'Not Available' values
for col in life_df.columns:
    full_row = life_df.loc[life_df[col] == 'Not Available', :]
    rows = full_row.iloc[:, 0:4]
    print(col)
    print(rows['Country'].describe())
    print()
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# frequency of null value for each year column (total for all nations)
nulls_list = []

for col in life_df.columns:
    full_row = life_df.loc[life_df[col] == 'Not Available', :]
    nulls_list.append(full_row[col].size)
    
print(nulls_list)
```

```python
fig1 = sns.barplot(x=list(life_df.columns),y=list(nulls_list))
fig1.set(xlabel='Years', ylabel='# Null Values', title = "Null frequency per year")
plt.xticks(rotation=90)
```
